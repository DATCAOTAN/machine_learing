### Import các thư viện

```python
import os
import pandas as pd
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
import sys
current_working_dir = os.getcwd()
project_path = os.path.abspath(os.path.join(current_working_dir, ".."))
project_path = project_path.replace("notebooks", "scripts")
print(project_path)

if project_path not in sys.path:
    sys.path.append(project_path)

from environment import TradingEnv
print("Imported TradingEnv successfully.")
```

###  Định nghĩa đường dẫn

```python

index = current_working_dir.find("trading_bot_rl_ppo")

# Cắt chuỗi đến vị trí kết thúc của "trading_bot_rl_ppo"
if index != -1:
    base_path = current_working_dir[:index + len("trading_bot_rl_ppo")]
else:
    base_path = current_working_dir  # Nếu không tìm thấy, giữ nguyên

model_path = os.path.join(base_path,r'models\ppo_trading_xauusd.zip')

```

### Tạo hoặc tải mô hình PPO

```python
if os.path.exists(model_path):
    model = PPO.load(model_path)
    print("Da load duoc model")
else:
    # Nếu mô hình chưa tồn tại, khởi tạo mới
    model = PPO("MlpPolicy", DummyVecEnv([lambda: TradingEnv(pd.DataFrame(), render_mode='human')]), verbose=1)
    print("dang khoi tao model")
```

### Tải dữ liệu và huấn luyện mô hình

```python
directory = r'data\processed\du_lieu_phan_tich'
for year in range(2015, 2016):
    for month in range(1,2):
        if month < 10:
            file_path = os.path.join(base_path,directory, str(year),'du_lieu_vang_phan_tich_'+str(year)+'_0'+str(month)+'.csv')
        else:
            file_path = os.path.join(base_path,directory,str(year),'du_lieu_vang_phan_tich_'+str(year)+'_'+str(month)+'.csv')

        # Đọc file CSV và kiểm tra sự tồn tại của file
        if os.path.exists(file_path):
            data = pd.read_csv(file_path)

            # Kiểm tra và xử lý các giá trị NaN trong dữ liệu

            # Tạo môi trường với dữ liệu hiện tại
            env = DummyVecEnv([lambda: TradingEnv(data, render_mode='human')])

            # Tiếp tục huấn luyện mô hình trên dữ liệu hiện tại
            timesteps = 1
            model.set_env(env)
            model.learn(total_timesteps=timesteps)

            print(f"Đã huấn luyện xong trên dữ liệu: {file_path}")
        else:
            print(f"File không tồn tại: {file_path}")
```

### Lưu mô hình đã huấn luyện

```python
# Lưu mô hình sau khi hoàn thành huấn luyện
model.save(model_path)
print(f"Mô hình đã được lưu tại: {model_path}")
```

###  Đánh giá nhanh hiệu suất mô hình

```python
# Đánh giá hiệu suất mô hình trên tệp cuối cùng
path_data=r"data\processed\du_lieu_phan_tich\2015\du_lieu_vang_phan_tich_2015_01.csv"
test_file = os.path.join(base_path,path_data)
print(test_file)
test_data = pd.read_csv(test_file)
test_env = DummyVecEnv([lambda: TradingEnv(test_data)])
model = PPO.load(model_path)

obs = test_env.reset()
total_reward = 0
done = False

while not done:
    action, _ = model.predict(obs, deterministic=True)
    print(f'action ={action}')
    obs, reward, done, info = test_env.step(action)
    profit = info[0].get('profit', 0)
    print(f'profit ={profit}')
    total_reward += profit
print(f"Tổng phần thưởng trên tập kiểm tra: {total_reward}")
```
