```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
data_path = r"C:\Users\nguye\OneDrive\documents\python\trading_bot_rl_ppo\data\raw\XAUUSD(2015-2024)\2015\XAUUSD_2015_01.csv"
data = pd.read_csv(data_path)

# Chuyển cột 'datetime' thành kiểu datetime và đặt làm chỉ mục
if 'datetime' in data.columns:
    data['datetime'] = pd.to_datetime(data['datetime'])
    data.set_index('datetime', inplace=True)

# Khám phá dữ liệu
print("Dữ liệu mẫu:")
print(data.head())

# Kiểm tra giá trị NaN
print("\nSố lượng giá trị thiếu trong mỗi cột:")
print(data.isnull().sum())

# Trực quan hóa giá đóng cửa
if 'close' in data.columns:
    data['close'].plot(figsize=(12, 6), title='Biểu đồ giá đóng cửa', ylabel='Close Price')
    plt.grid()
    plt.show()
else:
    print("'close' không tồn tại trong dữ liệu.")
```
